```python
import san
import pandas as pd
import quandl
import matplotlib.pyplot as plt
```

```python
gold=quandl.get("LBMA/GOLD", start_date="2019-01-01")["USD (PM)"]
gold.index = gold.index.tz_localize('UTC')
```

```python
btc = san.get('prices/bitcoin', from_date="2019-01-01")["priceUsd"]
```

```python
prices = pd.concat([gold, btc], axis=1).dropna()
```

```python
corr = prices["USD (PM)"].rolling(30).corr(prices["priceUsd"]).dropna()
```

```python
plt.figure(figsize=(20,10))
plt.xlabel("Date")
plt.ylabel("30 day rolling correlation")
plt.plot(corr)
plt.title("BTC vs Gold 30 day rolling correlation")
plt.axhline(color="red",ls="--")
plt.show()
```
